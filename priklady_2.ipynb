```python
from random import shuffle, choice, random
from reportlab.pdfgen.canvas import Canvas 
from reportlab.lib.units import cm, inch
from reportlab.lib.pagesizes import A4
from reportlab.lib import colors
from pathlib import Path

dst_fld_raw = r"c:\tmp\priklady"
dst_file_raw = "priklady_2x.pdf"
dst_path_raw = dst_fld_raw + "\\" + dst_file_raw
```

```python
operands = []
for i in range(3,10):
    for j in range(3,10):
        operands.append((i,j))

operands += operands
shuffle(operands)
operands.extend(operands[0:6]) # extra 6 pro zarovnani na 2 * 4 * 13
```

```python
data_operands = iter(operands)
data_operands_2 = []
while True:
    try:
        oper_a, oper_b = next(data_operands)
        oper_c, oper_d = next(data_operands)
        data_operands_2.append((oper_a, oper_b, oper_c, oper_d))
    except StopIteration:
        break
```

```python
priklady = []
for oper_a, oper_b, oper_c, oper_d in data_operands_2:
    mul_result_1 = oper_a * oper_b
    mul_result_2 = oper_c * oper_d    
    diff = mul_result_1 - mul_result_2

    priklad, result = "", ""
    if diff < -19 or diff in range(1,20):
        # prohod a odecti
        result = mul_result_2 - mul_result_1
        priklad = f"({oper_c} x {oper_d}) - ({oper_a} x {oper_b})"
    
    elif diff in range(-19,0):
        # odecti
        result = mul_result_1 - mul_result_2
        priklad = f"({oper_a} x {oper_b}) - ({oper_c} x {oper_d})"
    elif diff == 0 or diff >= 20 :
        # secti
        result = mul_result_1 + mul_result_2
        priklad = f"({oper_a} x {oper_b}) + ({oper_c} x {oper_d})"
    else:
        #print(oper_a, oper_b, oper_c, oper_d, diff) 
        print("Should not happen")
    
    priklad += " ="
    print (priklad, result)
    priklady.append((priklad, result))
```

```python
canvas = Canvas(dst_path_raw, pagesize=A4)

priklad_per_page = 13

v_position_start = 11 * inch
v_position_off = 0.8 * inch
v_position = v_position_start

canvas.setFont("Helvetica", 18)

counter = 1
for p in priklady:
    if counter > priklad_per_page:

        pgn = str(canvas.getPageNumber())
        canvas.setFont("Helvetica", 8)
        canvas.drawString(7*inch, 36, f"page: {pgn}")

        canvas.showPage()
        
        v_position_start = 11 * inch
        v_position_off = 0.8 * inch
        v_position = v_position_start
        
        canvas.setFont("Helvetica", 18)

        counter = 1

    txt = p[0]
    canvas.drawString(72, v_position, txt)
    #canvas.drawString(180, v_position, "= ")

    canvas.setStrokeColor(colors.lightgrey) 
    canvas.line(220,v_position - 8 ,290,v_position - 8)

    v_position -= v_position_off

    counter += 1

pgn = str(canvas.getPageNumber())
canvas.setFont("Helvetica", 8)
canvas.drawString(7*inch, 36, f"page: {pgn}")
canvas.showPage()
canvas.setFont("Helvetica", 18)

counter = 1
v_position_start = 11 * inch
v_position_off = 0.8 * inch
v_position = v_position_start
h_possition = inch
for p in priklady:
    if counter > priklad_per_page:
        h_possition += inch
        v_position_start = 11 * inch
        v_position_off = 0.8 * inch
        v_position = v_position_start
        counter = 1
    txt = str(p[1])
    canvas.drawString(h_possition, v_position, txt)
    v_position -= v_position_off
    counter += 1


canvas.save() 
```
