```python
from random import shuffle, choice, random
from reportlab.pdfgen.canvas import Canvas 
from reportlab.lib.units import cm, inch
from reportlab.lib.pagesizes import A4
from reportlab.lib import colors
from pathlib import Path

dst_fld = Path("c:/tmp/priklady")
dst_file = "priklady_9.pdf"
dst_path = dst_fld / dst_file
```

```python
def get_operands_pair(lo, hi):
    operands = []
    for i in range(lo, hi):
        for j in range(lo, hi):
            operands.append((i,j))
    shuffle(operands)
    while True:
        for operands_pair in operands:
            yield operands_pair
operands_pair_nact_nact = get_operands_pair(12, 19) # Create the operands_pair generator
operands_pair_low = get_operands_pair(5, 10) # Create the operands_pair generator low
```

```python
def find_divisors(dividend):
    divisors = []
    for i in range(2,10):
        if dividend % i == 0:
            divisors.append(i)
    return divisors

# find_divisors(21)
```

```python
def nasobeni_nact_nact(priklady):
    oper_a, oper_b = next(operands_pair_nact_nact)
    result = oper_a * oper_b
    if result > 330:
        return nasobeni_nact_nact(priklady) # try again
        
    priklad = f"{oper_a} x {oper_b}"
    priklady.append((priklad, result))         
    
```

```python
def deleni_zavorka(priklady, var=1):
    oper_a, oper_b = next(operands_pair_low)
  
    mul_result = oper_a * oper_b
    dividend = mul_result
    divisor = oper_a
    quotient = oper_b



    if var == 1:
        dividend_2 = choice([ i for i in range(3,10) if i != oper_b]) * divisor

        oper_left = dividend_2 - dividend
        oper_right = choice(list(range(19,39)))

        priklad = f"{oper_left} +  {dividend} ÷ {divisor} + {oper_right}"
        result = oper_left +  dividend // divisor + oper_right
        priklady.append((priklad, result)) 

        priklad = f"({oper_left} +  {dividend}) ÷ {divisor} + {oper_right}"
        result = (oper_left +  dividend) // divisor + oper_right
        priklady.append((priklad, result)) 

    else:

        oper_left = choice(list(range(5,39)))
        divisor_2 = choice(find_divisors(dividend))
        if divisor == divisor_2:
            return deleni_zavorka(priklady, var) # try again
        else:
            oper_right = divisor_2 - divisor
            
            if oper_left -  dividend // divisor + oper_right ==  oper_left -  dividend // (divisor + oper_right):
                return deleni_zavorka(priklady, var) # try again
            
            if dividend // divisor_2 > 10 :
                ... # napoveda pokud je vysledek deleni vetsi nez 10
                len_p  = len(priklady) % 13
                if len_p > 10: # napoveda na stejne strance
                    for _ in range(1, 13 - len_p + 1):
                        nasobeni_nact_nact(priklady)

                priklad = f"{divisor_2} x {dividend // divisor_2}"
                result = divisor_2 * (dividend // divisor_2)
                priklady.append((priklad, result)) 
        
        if oper_right < 0:
            oper_right *= -1

            priklad = f"{oper_left} -  {dividend} ÷ {divisor} - {oper_right}"
            result = oper_left -  dividend // divisor - oper_right
            priklady.append((priklad, result)) 

            priklad = f"{oper_left} -  {dividend} ÷ ({divisor} - {oper_right})"
            result = oper_left -  dividend // (divisor - oper_right)
            priklady.append((priklad, result)) 
        else:                        
            priklad = f"{oper_left} -  {dividend} ÷ {divisor} + {oper_right}"
            result = oper_left -  dividend // divisor + oper_right
            priklady.append((priklad, result)) 

            priklad = f"{oper_left} -  {dividend} ÷ ({divisor} + {oper_right})"
            result = oper_left -  dividend // (divisor + oper_right)
            priklady.append((priklad, result)) 
```

```python
priklady = []
while len(priklady) < (7*13):  # 13 prikladu na stranku, 7 stranek
  
        nasobeni_nact_nact(priklady)
        var = choice([1,2])
        deleni_zavorka(priklady, var=var)  

priklady = priklady[0:(7*13)] # ruzne funkce vraci ruzny pocet prikladu

for priklad, result in priklady:
    print(f"{priklad}= {result}")    

```

```python
canvas = Canvas(str(dst_path), pagesize=A4)

priklad_per_page = 13

v_position_start = 11 * inch
v_position_off = 0.8 * inch
v_position = v_position_start

canvas.setFont("Helvetica", 18)

counter = 1
for p in priklady:
    if counter > priklad_per_page:

        pgn = str(canvas.getPageNumber())
        canvas.setFont("Helvetica", 8)
        canvas.drawString(7*inch, 36, f"page: {pgn}")

        canvas.showPage()
        
        v_position_start = 11 * inch
        v_position_off = 0.8 * inch
        v_position = v_position_start
        
        canvas.setFont("Helvetica", 18)

        counter = 1

    txt = p[0]
    canvas.drawString(32, v_position, txt)
    canvas.drawString(190, v_position, "= ")

    canvas.setStrokeColor(colors.lightgrey) 
    canvas.line(220,v_position - 8 ,400,v_position - 8)
    canvas.setStrokeColor(colors.black) 
    canvas.drawString(430, v_position, "= ")
    canvas.setStrokeColor(colors.lightgrey) 
    canvas.line(460,v_position - 8 ,500,v_position - 8)


    v_position -= v_position_off

    counter += 1

pgn = str(canvas.getPageNumber())
canvas.setFont("Helvetica", 8)
canvas.drawString(7*inch, 36, f"page: {pgn}")
canvas.showPage()
canvas.setFont("Helvetica", 18)

counter = 1
v_position_start = 11 * inch
v_position_off = 0.8 * inch
v_position = v_position_start
h_possition = inch
for p in priklady:
    if counter > priklad_per_page:
        h_possition += inch
        v_position_start = 11 * inch
        v_position_off = 0.8 * inch
        v_position = v_position_start
        counter = 1
    txt = str(p[1])
    canvas.drawString(h_possition, v_position, txt)
    v_position -= v_position_off
    counter += 1


canvas.save()  
```
