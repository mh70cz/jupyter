```python
from random import shuffle, choice, random
from reportlab.pdfgen.canvas import Canvas 
from reportlab.lib.units import cm, inch
from reportlab.lib.pagesizes import A4
from reportlab.lib import colors
from pathlib import Path

dst_fld = Path("c:/tmp/priklady")
dst_file = "priklady_6.pdf"
dst_path = dst_fld / dst_file
```

```python
def get_operands_pair():
    operands = []
    for i in range(3,10):
        for j in range(3,10):
            operands.append((i,j))
    shuffle(operands)
    while True:
        for operands_pair in operands:
            yield operands_pair
operands_pair = get_operands_pair() # Create the operands_pair generator
```

```python
def get_operands_quad():
    o1 = choice([i for i in range(2,11)])
    o2 = choice([i for i in range(3,10)])

    if s := o1 + o2 <= 11:
        o3 = choice([i for i in range(4,10)])
    elif s == 12:
        o3 = choice([i for i in range(3,9)])   
    else:
        o3 = choice([i for i in range(2,5)])

    o4 = choice([i for i in range(51,100) if i % 10 != 0]) 

    return (o1, o2, o3, o4)         

```

```python
def nasobeni_zavorka(priklady, var=1):
    o1, o2,o3, o4 = get_operands_quad()

    if var == 1:
        priklad = f"{o1} + {o2} x {o3} + {o4}"
        result = o1 + o2 * o3 + o4
        priklady.append((priklad, result)) 

        priklad = f"({o1} + {o2}) x {o3} + {o4}"
        result = (o1 + o2) * o3 + o4
        priklady.append((priklad, result))
    else:
        priklad = f"{o4} + {o3} x {o2} + {o1}"
        result = o4 + o3 * o2 + o1
        priklady.append((priklad, result)) 

        priklad = f"{o4} + {o3} x ({o2} + {o1})"
        result = o4 + o3 * (o2 + o1)
        priklady.append((priklad, result))        
```

```python
def deleni(priklady, var=1):
    oper_a, oper_b = next(operands_pair)
  
    mul_result = oper_a * oper_b
    dividend = mul_result
    divisor = oper_a
    quotient = oper_b

    if var == 1:
        oper_c = choice(list(range(8,29)))
        priklad = f"{dividend} ÷ {divisor} - {oper_c}"
        result = quotient - oper_c
    else:
        oper_c = choice([i for i in range(39,100) if i % 10 != 0])
        priklad = f"{oper_c} + {dividend} ÷ {divisor}"
        result = quotient + oper_c
        
    priklady.append((priklad, result)) 
```

```python
def nasobeni(priklady, var=1):
    oper_a, oper_b = next(operands_pair)  
    mul_result = oper_a * oper_b
    if var == 1:
        if mul_result < 46:
            oper_c = mul_result + choice([i for i in range(39,54) if i % 10 != 0])
        elif mul_result < 54:
            oper_c = mul_result + choice([i for i in range(29,47) if i % 10 != 0])
        elif mul_result < 73:
            oper_c = mul_result + choice([i for i in range(19,30) if i % 10 != 0])
        else:
            oper_c = mul_result + choice(list(range(13,20)))

        priklad = f"{oper_c} - {oper_a} x {oper_b}"
        result =  oper_c - mul_result
    else:
        oper_c = choice([i for i in range(29,80) if i % 10 != 0])
        priklad = f"{oper_a} x {oper_b} + {oper_c}"
        result =  oper_c + mul_result 
        
    priklady.append((priklad, result))     
```

```python
def odcitani_zavorka(priklady, var=1):
    if var == 1:
        oper_a = choice(list(range(69,100)))
        oper_b = choice([i for i in range(29,oper_a - 20) if i % 10 != 0])
        oper_c = choice([i for i in range(11,oper_a - oper_b) if (i % 10 != 0 and i < oper_b)])

        priklad = f"{oper_a} - {oper_b} - {oper_c}"
        result = oper_a - oper_b - oper_c
        priklady.append((priklad, result)) # extra append

        priklad = f"{oper_a} - ({oper_b} - {oper_c})"
        result = oper_a - (oper_b - oper_c)
        priklady.append((priklad, result))
    else:
        oper_a = choice(list(range(69,100)))
        oper_b = choice([i for i in range(29,oper_a - 20) if i % 10 != 0])
        oper_c = choice([i for i in range(11,oper_a - oper_b) if (i % 10 != 0 and i < oper_b)])

        priklad = f"{oper_a} - {oper_b} + {oper_c}"
        result = oper_a - oper_b + oper_c
        priklady.append((priklad, result)) # extra append

        priklad = f"{oper_a} - ({oper_b} + {oper_c})"
        result = oper_a - (oper_b + oper_c)
        priklady.append((priklad, result))     
```

```python
priklady = []
while len(priklady) < (7*13):  # 13 prikladu na stranku, 7 stranek
    nasobeni_zavorka(priklady, 1)
    deleni(priklady, 1)
    odcitani_zavorka(priklady, 1)
    nasobeni(priklady, 2)
    nasobeni_zavorka(priklady, 2)
    deleni(priklady, 2)
    odcitani_zavorka(priklady, 2)    
    nasobeni(priklady, 1)

priklady = priklady[0:(7*13)] # ruzne funkce vraci ruzny pocet prikladu


for priklad, result in priklady:
    print(f"{priklad}= {result}")       
```

```python
canvas = Canvas(str(dst_path), pagesize=A4)

priklad_per_page = 13

v_position_start = 11 * inch
v_position_off = 0.8 * inch
v_position = v_position_start

canvas.setFont("Helvetica", 18)

counter = 1
for p in priklady:
    if counter > priklad_per_page:

        pgn = str(canvas.getPageNumber())
        canvas.setFont("Helvetica", 8)
        canvas.drawString(7*inch, 36, f"page: {pgn}")

        canvas.showPage()
        
        v_position_start = 11 * inch
        v_position_off = 0.8 * inch
        v_position = v_position_start
        
        canvas.setFont("Helvetica", 18)

        counter = 1

    txt = p[0]
    canvas.drawString(72, v_position, txt)
    canvas.drawString(220, v_position, "= ")

    canvas.setStrokeColor(colors.lightgrey) 
    canvas.line(250,v_position - 8 ,400,v_position - 8)
    canvas.setStrokeColor(colors.black) 
    canvas.drawString(430, v_position, "= ")
    canvas.setStrokeColor(colors.lightgrey) 
    canvas.line(460,v_position - 8 ,500,v_position - 8)


    v_position -= v_position_off

    counter += 1

pgn = str(canvas.getPageNumber())
canvas.setFont("Helvetica", 8)
canvas.drawString(7*inch, 36, f"page: {pgn}")
canvas.showPage()
canvas.setFont("Helvetica", 18)

counter = 1
v_position_start = 11 * inch
v_position_off = 0.8 * inch
v_position = v_position_start
h_possition = inch
for p in priklady:
    if counter > priklad_per_page:
        h_possition += inch
        v_position_start = 11 * inch
        v_position_off = 0.8 * inch
        v_position = v_position_start
        counter = 1
    txt = str(p[1])
    canvas.drawString(h_possition, v_position, txt)
    v_position -= v_position_off
    counter += 1


canvas.save()   
```
