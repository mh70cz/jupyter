```python
from random import shuffle, choice, random
from reportlab.pdfgen.canvas import Canvas 
from reportlab.lib.units import cm, inch
from reportlab.lib.pagesizes import A4
from reportlab.lib import colors
from pathlib import Path

dst_fld = Path("c:/tmp/priklady")
dst_file = "priklady_8.pdf"
dst_path = dst_fld / dst_file
```

```python
def get_operands_pair(lo, hi):
    operands = []
    for i in range(lo, hi):
        for j in range(lo, hi):
            operands.append((i,j))
    shuffle(operands)
    while True:
        for operands_pair in operands:
            yield operands_pair
operands_pair = get_operands_pair(11, 19) # Create the operands_pair generator
operands_pair_low = get_operands_pair(5, 10) # Create the operands_pair generator low
```

```python

def sub_priklad(oper_a, oper_b):
    mul_result = oper_a * oper_b

    if mul_result > 200:
        return None 

    priklad = f"{10} x {oper_b} + {oper_a - 10} x {oper_b}"
    result =  mul_result
    return (priklad, result)
```

```python
def deleni(priklady, oper_a, oper_b):
    var = choice([1,2,3,4])
    mul_result = oper_a * oper_b
    dividend = mul_result
    divisor = oper_a
    quotient = oper_b    

    if var == 1:
        oper_c = choice([i for i in range(8,30) if i % 10 != 0])
        priklad = f"{dividend} ÷ {divisor} - {oper_c}"
        result = quotient - oper_c
    elif var == 2:
        oper_c = choice([i for i in range(39,100) if i % 10 != 0])
        priklad = f"{dividend} ÷ {divisor} + {oper_c}"
        result = quotient + oper_c
    elif var == 3:
        oper_c = choice([i for i in range(-19,19) if i % 10 != 0])
        priklad = f"{oper_c} - {dividend} ÷ {divisor}"
        result = oper_c - quotient                               
    else:
        oper_c = choice([i for i in range(39,100) if i % 10 != 0])
        priklad = f"{oper_c} + {dividend} ÷ {divisor}"
        result = oper_c + quotient
        
    priklady.append((priklad, result))    
```

```python
priklady = []
while len(priklady) < (7*13):  # 13 prikladu na stranku, 7 stranek
    while True:
        oper_a, oper_b = next(operands_pair)
        if (pr1 := sub_priklad(oper_a, oper_b)) is None:
            continue
        priklady.append(pr1)
        oper_a1, oper_b1 = oper_a, oper_b 
        break
    while True:
        oper_a, oper_b = next(operands_pair)
        if (pr2 := sub_priklad(oper_a, oper_b)) is None:
            continue
        priklady.append(pr2)
        oper_a2, oper_b2 = oper_a, oper_b 
        break

    deleni(priklady, *next(operands_pair_low))

    if oper_a1 * oper_b1 + oper_a2 * oper_b2 < 300:
        priklad = f"{oper_a1} x {oper_b1} + {oper_a2} x {oper_b2}"
        result = oper_a1 * oper_b1 + oper_a2 * oper_b2
    elif oper_a1 * oper_b1 > oper_a2 * oper_b2:
        priklad = f"{oper_a1} x {oper_b1} - {oper_a2} x {oper_b2}"
        result = oper_a1 * oper_b1 - oper_a2 * oper_b2
    else:
        priklad = f"{oper_a2} x {oper_b2} - {oper_a1} x {oper_b1}"
        result = oper_a2 * oper_b2 - oper_a1 * oper_b1  

    priklady.append((priklad, result))

    if (len(priklady) + 1)  % 13 == 0:
        deleni(priklady, *next(operands_pair_low))

priklady = priklady[0:(7*13)] # ruzne funkce vraci ruzny pocet prikladu

for priklad, result in priklady:
    print(f"{priklad} = {result}")
```

```python
canvas = Canvas(str(dst_path), pagesize=A4)

priklad_per_page = 13

v_position_start = 11 * inch
v_position_off = 0.8 * inch
v_position = v_position_start

canvas.setFont("Helvetica", 18)

counter = 1
for p in priklady:
    if counter > priklad_per_page:

        pgn = str(canvas.getPageNumber())
        canvas.setFont("Helvetica", 8)
        canvas.drawString(7*inch, 36, f"page: {pgn}")

        canvas.showPage()
        
        v_position_start = 11 * inch
        v_position_off = 0.8 * inch
        v_position = v_position_start
        
        canvas.setFont("Helvetica", 18)

        counter = 1

    txt = p[0]
    canvas.drawString(32, v_position, txt)
    canvas.drawString(190, v_position, "= ")

    canvas.setStrokeColor(colors.lightgrey) 
    canvas.line(220,v_position - 8 ,400,v_position - 8)
    canvas.setStrokeColor(colors.black) 
    canvas.drawString(430, v_position, "= ")
    canvas.setStrokeColor(colors.lightgrey) 
    canvas.line(460,v_position - 8 ,500,v_position - 8)


    v_position -= v_position_off

    counter += 1

pgn = str(canvas.getPageNumber())
canvas.setFont("Helvetica", 8)
canvas.drawString(7*inch, 36, f"page: {pgn}")
canvas.showPage()
canvas.setFont("Helvetica", 18)

counter = 1
v_position_start = 11 * inch
v_position_off = 0.8 * inch
v_position = v_position_start
h_possition = inch
for p in priklady:
    if counter > priklad_per_page:
        h_possition += inch
        v_position_start = 11 * inch
        v_position_off = 0.8 * inch
        v_position = v_position_start
        counter = 1
    txt = str(p[1])
    canvas.drawString(h_possition, v_position, txt)
    v_position -= v_position_off
    counter += 1


canvas.save()  
```
