```python
from random import shuffle, choice, random
from reportlab.pdfgen.canvas import Canvas 
from reportlab.lib.units import cm, inch
from reportlab.lib.pagesizes import A4
from reportlab.lib import colors
from pathlib import Path

dst_fld = Path("c:/tmp/priklady")
dst_file = "priklady_4.pdf"
dst_path = dst_fld / dst_file
```

```python
def get_random_single_used_value(values):
    if values:  # Check if the list is not empty
        # Select a random value from the list
        selected_value = random.choice(values)
        # Remove the selected value from the list
        values.remove(selected_value)
        return selected_value
    else:
        return None  # Return None if the list is empty
    

def toggle_values():
    values = [1, 2, 3, 4]  # List of values to toggle between
    while True:
        for value in values:
            yield value    
# Create the toggle generator
toggle = toggle_values()            
```

```python
operands = []
for i in range(3,10):
    for j in range(3,10):
        operands.append((i,j))

shuffle(operands)
operands.extend(operands[0:3]) # extra 3 pro zarovnani na 4 * 13
```

```python
# minus_results = list(range(-11,-1))
# shuffle(minus_minus_results)
numbers_for_oper_c = list(range(-21, -5)) + list(range(31, 100))
```

```python
priklady = []
for o in operands:
    
    oper_a, oper_b = o
   #  oper_c = get_random_single_used_value(numbers_for_oper_c )

    mul_result = oper_a * oper_b
    dividend = mul_result
    divisor = oper_a
    quotient = oper_b

    priklad = result = None

    toggle_value = next(toggle)
    if toggle_value == 1:
        oper_c = choice(list(range(8,29)))
        priklad = f"{dividend} ÷ {divisor} - {oper_c}"
        result = quotient - oper_c
    elif toggle_value == 2:
        oper_c = choice(list(range(-9, -3)) + list(range(31, 100)))
        priklad = f"{oper_c} - {dividend} ÷ {divisor}"
        result =  oper_c - quotient
    elif toggle_value == 3:
        oper_c = choice(list(range(13,39)))
        priklad = f"{dividend} ÷ {divisor} + {oper_c}"
        result = quotient + oper_c
    elif toggle_value == 4:
        oper_c = choice(list(range(-21, -5)) + list(range(31, 80)))
        priklad = f"{oper_c} + {dividend} ÷ {divisor}"
        result =  oper_c + quotient                 

    else:
        pass #should not happen

    print(f"{priklad}= {result}")
    priklady.append((priklad, result)) 

```

```python
canvas = Canvas(str(dst_path), pagesize=A4)

priklad_per_page = 13

v_position_start = 11 * inch
v_position_off = 0.8 * inch
v_position = v_position_start

canvas.setFont("Helvetica", 18)

counter = 1
for p in priklady:
    if counter > priklad_per_page:

        pgn = str(canvas.getPageNumber())
        canvas.setFont("Helvetica", 8)
        canvas.drawString(7*inch, 36, f"page: {pgn}")

        canvas.showPage()
        
        v_position_start = 11 * inch
        v_position_off = 0.8 * inch
        v_position = v_position_start
        
        canvas.setFont("Helvetica", 18)

        counter = 1

    txt = p[0]
    canvas.drawString(72, v_position, txt)
    canvas.drawString(180, v_position, "= ")

    canvas.setStrokeColor(colors.lightgrey) 
    canvas.line(220,v_position - 8 ,370,v_position - 8)
    canvas.setStrokeColor(colors.black) 
    canvas.drawString(390, v_position, "= ")
    canvas.setStrokeColor(colors.lightgrey) 
    canvas.line(430,v_position - 8 ,470,v_position - 8)


    v_position -= v_position_off

    counter += 1

pgn = str(canvas.getPageNumber())
canvas.setFont("Helvetica", 8)
canvas.drawString(7*inch, 36, f"page: {pgn}")
canvas.showPage()
canvas.setFont("Helvetica", 18)

counter = 1
v_position_start = 11 * inch
v_position_off = 0.8 * inch
v_position = v_position_start
h_possition = inch
for p in priklady:
    if counter > priklad_per_page:
        h_possition += inch
        v_position_start = 11 * inch
        v_position_off = 0.8 * inch
        v_position = v_position_start
        counter = 1
    txt = str(p[1])
    canvas.drawString(h_possition, v_position, txt)
    v_position -= v_position_off
    counter += 1


canvas.save()    
```
